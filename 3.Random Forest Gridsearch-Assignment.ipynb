```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# open a Dataset
```

```python
dataset=pd.read_csv("insurance_pre.csv")
```

```python
dataset
```

```python
#Remove the categorical data (i.e str change into numbers)
```

```python
dataset= pd.get_dummies(dataset,dtype=int,drop_first=True)
```

```python
dataset
```

```python
# Split the Input  Output dataes
```

```python
dataset.columns
```

```python
independant=dataset[['age','bmi', 'children','sex_male','smoker_yes']]
                     
```

```python
independant
```

```python
dependant= dataset[['charges']]
```

```python
dependant
```

```python
# Model creation from Test data
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor
param_grid = {'criterion':['squarded_error','absolute_error','poission''mae','friendman_mse'],'max_features':['auto','sqrt','log2'],'n_estimators':[10,100]}

grid= GridSearchCV(RandomForestRegressor(),param_grid,refit=True,verbose=3,n_jobs=-1)

# Fitting the Model for grid seach                 
grid.fit(independant,dependant)
```

```python
# print best parameter after turning
# print(grid_best_params_)
```

```python
re=grid.cv_results_
print("The r_score_value for best parameters:",format(grid.best_params_))
```

```python
table=pd.DataFrame.from_dict(re)
```
